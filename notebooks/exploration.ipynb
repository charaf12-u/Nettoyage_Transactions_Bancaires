```python
import pandas as pd

# --> read data in csv
file_path = "../data/bank-transactions.csv"
df = pd.read_csv(file_path)

print("Dataset loaded successfully")
```

```python
# --> afficher les valeurs unique pour devise et segment_clients
print("\nDevise unique:")
print(df["devise"].unique())

print("\nSegment client unique:")
print(df["segment_client"].unique())
```

```python
import matplotlib.pyplot as plt

# --> afficher distribution des montants par un graphe
montants = df["montant"].astype(str).str.replace(",", ".").astype(float)

plt.hist(montants, bins=30)
plt.title("Distribution des montants")
plt.show()
```

```python
# --> calculs nomber des valeurs null
missing_values = df.isnull().sum()

# --> calculs le porsentage pour les valeurs null
missing_percentage = (missing_values / len(df)) * 100

# --> afficher les valeurs pour un table
missing_table = pd.DataFrame({
    "missing_values": missing_values,
    "missing_percentage": missing_percentage
})

# --> afficher resultat
missing_table[missing_table["missing_values"] > 0]
```

```python

# --> afficher le nomber des valeur null in columns pour un graphes
missing_values.plot(kind="bar")

plt.title("Nombre de valeurs manquantes par colonne")
plt.xlabel("Colonnes")
plt.ylabel("Nombre de valeurs manquantes")
plt.xticks(rotation=45)
plt.show()
```

```python
# --> calculs le nomber des transaction_id duplicated
print("le nomber duplicated :",df.duplicated(subset="transaction_id").sum())
```

```python
# --> afficher les formation pour les duplicate de transaction_id
duplicates = df[df.duplicated(subset="transaction_id", keep=False)]
duplicates.sort_values("transaction_id")

duplicates[[
    "transaction_id",
    "client_id",
    "montant",
    "devise",
    "date_transaction"
]].sort_values("transaction_id")
```
